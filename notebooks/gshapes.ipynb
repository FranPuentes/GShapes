```python
import requests
from PIL import Image
from io import BytesIO

def cargar_imagen(url):
    respuesta = requests.get(url)
    imagen = Image.open(BytesIO(respuesta.content))
    return imagen;

# Ejemplo de uso
url = "https://d138zd1ktt9iqe.cloudfront.net/media/seo_landing_files/file-shapes-for-kids-1592568510.jpg"
imagen = cargar_imagen(url)

display(imagen);
```

```python
from ultralytics import YOLO;

model = YOLO("./best.pt");

results = model(imagen, conf=0.80);

for result in results:
    tmp = result.plot();
    display(tmp);
```
